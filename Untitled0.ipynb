```python
import pandas as pd
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
```

```python
train_data = pd.read_csv(r"/content/samsum-train.csv")
val_data = pd.read_csv(r"/content/samsum-validation.csv")
```

Training and Validation Data

```python
train_data = train_data.sample(n = 4000, random_state = 42).reset_index(drop = True)
val_data = val_data.sample(n = 500, random_state = 42).reset_index(drop = True)
```

```python
import re

def clean_data(text):
  text = re.sub(r"\r\n", ' ', text)
  text = re.sub(r"<.*?", ' ', text)
  text = re.sub(r"\s+", ' ', text)
  text = text.strip().lower()
  return text
```

Data PreProcessing

```python
train_data['dialogue'] = train_data['dialogue'].apply(clean_data)
train_data['summary'] = train_data['summary'].apply(clean_data)

val_data['dialogue'] = val_data['dialogue'].apply(clean_data)
val_data['summary'] = val_data['summary'].apply(clean_data)
```

Tokenisation

```python
tokeniser = T5Tokenizer.from_pretrained("T5-small")
```

```python
def tokenise(data):
  inputs = tokeniser(data['dialogue'], padding = "max_length", max_length = 512, truncation = True)
  target = tokeniser(data['summary'], padding = "max_length", max_length = 200, truncation = True)

  inputs['labels'] = target["input_ids"]
  return inputs
```

```python
train_dataset = train_data.apply(tokenise, axis = 1).tolist()
val_dataset = val_data.apply(tokenise, axis = 1).tolist()
```

Working with our Model

```python
model = T5ForConditionalGeneration.from_pretrained("T5-small")
```

```python
import torch

if torch.cuda.is_available():
  device = torch.device("cuda")
else:
  device = torch.device("cpu")
```

```python
 model.to(device)
```

Training Arguments

```python
train_args = TrainingArguments(
    output_dir = './results',
    num_train_epochs = 10,
    weight_decay = 0.01,

    per_device_train_batch_size = 8,
    per_device_eval_batch_size = 8,

    eval_strategy = 'epoch',
    save_strategy = 'epoch',

    warmup_steps= 500

)
```

```python
trainer = Trainer(
    model = model,
    args = train_args,
    train_dataset = train_dataset,
    eval_dataset = val_dataset
)
```

```python
trainer.train()
```

```python
model.save_pretrained('./saved_model_summary')
tokeniser.save_pretrained('./saved_model_summary')
#
```

```python
model= T5ForConditionalGeneration.from_pretrained('./saved_model_summary')
tokeniser = T5Tokenizer.from_pretrained('./saved_model_summary')
```

Testing

```python
def summarise_dialogue(dialogue):
  #clean
  dialogue = clean_data(dialogue)

  #tokenize
  inputs = tokeniser(
      dialogue, padding = "max_length", max_length = 512, truncation = True, return_tensors = "pt"
  ).to(device)

  model.to(device)

  #generate summary
  target = model.generate(
      input_ids = inputs['input_ids'],
      attention_mask = inputs['attention_mask'],
      max_length = 150, num_beams = 4,
      early_stopping = True
                          )
  #decoded output
  summary = tokeniser.decode(target[0], skip_special_tokens = True)

  return summary
```

```python
test_dialogue = """Sarah: Hey Mark, are you free this weekend? I really need a break from work.
Mark: Same here! I was thinking of going hiking at Green Mountain, but I’m open to ideas.
Sarah: That sounds perfect! I’ve been wanting to see the views there. Should we pack lunch?
Mark: Definitely. I can bring some snacks and a first-aid kit, just in case.
Sarah: Great, I’ll make sandwiches and bring enough water for both of us.
Mark: Do you think we should leave early to avoid the midday heat?
Sarah: Good point. Let’s meet at my place around 7:00 AM on Saturday.
Mark: Sounds like a solid plan. I’m looking forward to it!"""

summary = summarise_dialogue(test_dialogue)
print(summary)
```
